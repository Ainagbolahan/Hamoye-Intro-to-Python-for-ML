Which of the following fuel type code has the lowest average fuel cost per unit burned?

```python
fig = px.scatter(df, 
                 x='fuel_type_code_pudl', 
                 y='fuel_cost_per_unit_burned', 
                 color='report_year', 
                 opacity=0.8, 
                 hover_data=['report_year'], 
                 title='Fuel type code to Fuel cost')
fig.update_traces(marker_size=5)
fig.show()

```

```python
coal_df = df.loc[df.fuel_type_code_pudl == 'coal']
```

```python
coal_df = coal_df.groupby('report_year')['fuel_cost_per_unit_burned'].sum()
```

# Question 10
Which year has the highest average fuel cost per unit delivered?

```python
fig = px.scatter(df, 
                 x='report_year', 
                 y='fuel_cost_per_unit_delivered', 
                 color='report_year', 
                 opacity=0.8, 
                 hover_data=['report_year'], 
                 title='Report_year/Cost_Delivered')
fig.update_traces(marker_size=5)
fig.show()
```

```python
df.groupby('report_year')['fuel_cost_per_unit_delivered'].mean().sort_values(ascending=False)
```
